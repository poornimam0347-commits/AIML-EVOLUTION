```python
df = pd.read_excel("Combined_Land_Investment_Dataset_Single_Sheet.xlsx")
df = df.loc[:, ~df.columns.str.contains("Unnamed")]
df.head()
```

```python
category_map = {
    "Low": 1,
    "Mid": 2,
    "High": 3,
    "Very High": 4
}
df["Investment_Level"] = df["Investment Category"].map(category_map)
df["Score_Normalized"] = df["Investment Score(1-10)"] / 10
df.head()
```

```python
years = list(range(2024, 2031))
forecast_data = []

for _, row in df.iterrows():
    base_price = row["Avg Price per sq.ft (₹)"]
    growth_rate = row["Score_Normalized"] * 0.12  # intelligence-based growth
    
    price = base_price
    for year in years:
        price = price * (1 + growth_rate)
        forecast_data.append([
            row["Area / Locality"],
            row["Location"],
            year,
            round(price, 2)
        ])

forecast_df = pd.DataFrame(
    forecast_data,
    columns=["Area", "Location", "Year", "Predicted Price"]
)

forecast_df.head()
```

```python
plt.figure()
plt.scatter(
    df["Avg Price per sq.ft (₹)"],
    df["Investment Score(1-10)"]
)
plt.xlabel("Average Land Price (₹/sq.ft)")
plt.ylabel("Investment Score")
plt.title("Investment Impact Map: Price vs Intelligence Score")
plt.show()
```

```python
blr = df[df["Location"] == "Bangalore"]
non_blr = df[df["Location"] != "Bangalore"]
plt.figure()
plt.bar(
    ["Bangalore (Business Use)", "Outside Bangalore (Investment Use)"],
    [
        blr["Investment Score(1-10)"].mean(),
        non_blr["Investment Score(1-10)"].mean()
    ]
)

plt.ylabel("Average Investment Score")
plt.title("Business vs Investment Land Analysis")
plt.show()
```

```python
top_areas = forecast_df["Area"].unique()[:4]
plt.figure()
for area in top_areas:
    area_data = forecast_df[forecast_df["Area"] == area]
    plt.plot(area_data["Year"], area_data["Predicted Price"], label=area)
plt.xlabel("Year")
plt.ylabel("Predicted Price (₹/sq.ft)")
plt.title("AI-Driven Land Price Forecast (2024–2030)")
plt.legend()
plt.show()
```

```python
def investment_decision(score):
    if score >= 8.2:
        return "Strong Invest"
    elif score >= 7.5:
        return "Moderate Invest"
    else:
        return "Avoid"

df["Investment Decision"] = df["Investment Score(1-10)"].apply(investment_decision)

df[["Area / Locality", "Investment Score(1-10)", "Investment Decision"]].head(10)
```
